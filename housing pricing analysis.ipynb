```python
#outliers

def Outliers(columns):
    outlier = []
    
    Q1 = columns.quantile(.25)
    Q3 = columns.quantile(.75)
    
    IQR = Q3 - Q1
    lower_limits = Q1-(1.5 * IQR)
    upper_limits = Q3 + (1.5 * IQR)
    
    for out in columns:
        if out < lower_limits or out > upper_limits:
            outlier.append(out)
            
    return np.array(outlier)
    
    
```

```python
#show the number of outliers in the data

print(len(Outliers(file_data.price)))
print(len(Outliers(file_data.new_price)))
print(len(Outliers(file_data.bedrooms)))
print(len(Outliers(file_data.bathrooms)))
print(len(Outliers(file_data.toilets)))
print(len(Outliers(file_data.estate)))
print(len(Outliers(file_data.terrace)))
print(len(Outliers(file_data.location_rank)))
print(len(Outliers(file_data.estate_price)))
print(len(Outliers(file_data.serviced_price)))


```

```python
#removing the outliers
def removeOutliers(column):
    sorted(column)
    Q1,Q3 = column.quantile([0.25,0.75])
    IQR = Q3-Q1
    lower_limits = Q1-(1.5*IQR)
    upper_limits = Q3+(1.5* IQR)
    return lower_limits , upper_limits

    
```

```python
#removing data outliers in price data detected
low_price,high_price = removeOutliers(file_data["price"])
file_data["price"]=np.where(file_data["price"] > high_price , high_price , file_data["price"])
file_data["price"]=np.where(file_data["price"] < low_price , low_price , file_data["price"])
```

```python
#removing data outliers in estate price data detected
low_estate_price,high_estate_price = removeOutliers(file_data["estate_price"])
file_data["estate_price"]=np.where(file_data["estate_price"] > high_estate_price , high_estate_price , file_data["estate_price"])
file_data["estate_price"]=np.where(file_data["estate_price"] < low_estate_price , low_estate_price , file_data["estate_price"])
```
