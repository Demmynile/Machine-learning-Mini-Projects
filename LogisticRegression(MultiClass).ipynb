```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_digits
from sklearn.linear_model import LogisticRegression
```

```python
digits = load_digits()
dir(digits)
```

```python
digits.data[0]
```

```python
%matplotlib inline
for i in range(5):
    
    plt.gray()
    plt.matshow(digits.images[i])
    plt.gcf().set_size_inches(16,9)
```

```python
#checking out target and target names
digits.target[0:5]
```

```python
digits.target_names
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(digits.data , digits.target , test_size = 0.2)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train , y_train)
```

```python
 model.score(x_test , y_test)
```

```python
plt.matshow(digits.images[75])
```

```python
digits.target[75]
```

```python
model.predict(digits.data[0:5])
```

```python
#confusion matrix
#this will help us know where our models didnt do well
y_predicted = model.predict(x_test)
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test , y_predicted)
cm
```

```python
import seaborn as sns
sns.heatmap(cm , annot = True)
plt.gcf().set_size_inches(16,9)
plt.xlabel("predicted")
plt.ylabel("trurh")
```

```python
#create a logistics model
model_iris = LogisticRegression()
```

```python
#using train_test_split method
x_train , x_test , y_train, y_test = train_test_split(iris.data , iris.target , test_size = 0.2)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
model_iris.fit(x_train, y_train)
```

```python
model_iris.predict(x_test)
```

```python
y_test
```

```python
model_iris.score(x_train , y_train)
```

```python
model_iris.score(x_test , y_test)
```

```python
#to check for the confusion matrix
from sklearn.metrics import confusion_matrix
```

```python
y_predicted = model_iris.predict(x_test)
cm = confusion_matrix(y_predicted , y_test)
```

```python
import seaborn as sns
```

```python
sns.heatmap(cm , annot = True)
plt.xlabel("predicted")
plt.ylabel("truth")
plt.gcf().set_size_inches(16,9)
```
